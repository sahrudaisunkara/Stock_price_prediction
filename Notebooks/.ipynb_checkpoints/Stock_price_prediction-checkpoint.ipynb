```python
import numpy as np 
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
import os
os.makedirs("models", exist_ok=True)
os.makedirs("scalers", exist_ok=True)
```

```python
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import joblib
```

```python
stocks = {
    "INFY.NS": "Infosys",
    "TCS.NS": "TCS",
    "RELIANCE.NS": "Reliance",
    "HDFCBANK.NS": "HDFC Bank",
    "ICICIBANK.NS": "ICICI Bank",
    "ITC.NS": "ITC",
    "SBIN.NS": "SBI",
    "LT.NS": "L&T",
    "AAPL": "Apple",
    "MSFT": "Microsoft"
}
```

```python
WINDOW = 60

def fetch_data(ticker):
    df = yf.download(ticker, start="2018-01-01")
    df.reset_index(inplace=True)
    df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
    df.dropna(inplace=True)
    return df


def create_sequences(X_scaled, y_scaled, window=60):
    X, y = [], []
    for i in range(window, len(X_scaled)):
        X.append(X_scaled[i-window:i])
        y.append(y_scaled[i])
    return np.array(X), np.array(y)


def build_model():
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(WINDOW, 5)),
        Dropout(0.2),
        LSTM(32),
        Dropout(0.2),
        Dense(2)   # High, Low
    ])
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
for ticker, name in stocks.items():
    print(f"\n Training model for {name} ({ticker})")

    #  Fetch data
    df = fetch_data(ticker)
    print("Rows:", len(df))

    #  Prepare inputs & targets
    X_features = ['Open', 'High', 'Low', 'Close', 'Volume']
    y_targets = ['High', 'Low']

    X_data = df[X_features].values
    y_data = df[y_targets].values

    #  Scale separately
    X_scaler = MinMaxScaler()
    y_scaler = MinMaxScaler()

    X_scaled = X_scaler.fit_transform(X_data)
    y_scaled = y_scaler.fit_transform(y_data)

    #  Create sequences
    X, y = create_sequences(X_scaled, y_scaled, WINDOW)

    # Safety check
    if len(X) < 100:
        print("âš ï¸ Not enough data, skipping")
        continue

    # Train-test split
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    #  Build & train model
    model = build_model()

    model.fit(
        X_train,
        y_train,
        epochs=10,
        batch_size=32,
        validation_data=(X_test, y_test),
        verbose=0
    )

    #  Save model & scalers
    model.save(f"models/{ticker}_model.h5")
    joblib.dump(X_scaler, f"scalers/{ticker}_X_scaler.pkl")
    joblib.dump(y_scaler, f"scalers/{ticker}_y_scaler.pkl")

    print(f"âœ… Saved model & scalers for {name}")
```
